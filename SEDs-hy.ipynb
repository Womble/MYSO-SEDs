```hy
(def *python-slice* (get --builtins-- "slice"))
(defreader | [expr] ;reader macro for simple construction of complex slices, designed for use on numpy arrays
  `(get ~(car expr) (, ~@(list-comp (cond [(= (type x) HyExpression) (+ '(*python-slice*) x)]
                                          [(= x "None")  '(*python-slice* 0 None)]
                                          [True x])
                                  [x (cdr expr)]))))      
```
